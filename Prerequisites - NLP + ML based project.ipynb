# Lemmatization by using WordNetLemmatizer

```python
from nltk.stem import WordNetLemmatizer
wl = WordNetLemmatizer()
# pos: The Part Of Speech tag. Valid options are `"n"` for nouns,
#     `"v"` for verbs, `"a"` for adjectives, `"r"` for adverbs and `"s"`
#     for satellite adjectives.
print(wl.lemmatize('history', pos = 'v'))
print(wl.lemmatize('congratulations', pos = 'v'))
print(wl.lemmatize('tasty', pos = 'v'))
print(wl.lemmatize('tasteful', pos = 'v'))
print(wl.lemmatize('tasteless', pos = 'v'))
print(wl.lemmatize('delicious', pos = 'v'))
```

# Perform Text Pre-processing on smaller DataFrame to check if it is working correctly or not

## Make sure that the  smaller DataFrame also has a mixture of both - true job postings and fake job postings. Otherwise, there would be an error while we try to fit the ML model on training dataset.

```python
df = jobs.iloc[0:10]
df
```

```python
import re
corpus = []
for i in range(len(df)):
    #Remove Hyperlinks
    job_posting = re.sub(r"http\S+", ' ', df['text'][i])    
    #job_posting = re.sub(r'https?:\/\/\S*', '', df['text'][i], flags=re.MULTILINE)
    
    #Remove Punctuation Marks and Special Symbols
    job_posting = re.sub('[^a-zA-Z0-9]', ' ', df['text'][i])
    
    #Lowercase
    job_posting = job_posting.lower()
    
    #Create a list of strings using string.split() method
    job_posting = job_posting.split()
    
    
    #job_posting = [ps.stem(word) for word in job_posting if not word in stopwords.words('english')]
    job_posting = [wl.lemmatize(word, pos='v') for word in job_posting if not word in stopwords.words('english')]    
    job_posting = ' '.join(job_posting)
    corpus.append(job_posting)
    print(i, end = ' ')
```

```python
df['processed_text'] = corpus
df
```
